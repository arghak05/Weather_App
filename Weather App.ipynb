```python
import requests
from tkinter import *
from tkinter import messagebox
from PIL import ImageTk,Image
import tkinter as tk

api_key = 'cbf40c31dc6e53a5b63069b956572b3a'
url = 'http://api.openweathermap.org/data/2.5/weather?q={}&appid={}'


def getweather(city):
    result = requests.get(url.format(city, api_key))

    if result:
        json = result.json()
        city = json['name']
        country = json['sys']
        temp_kelvin = json['main']['temp']
        temp_celsius = temp_kelvin-273.15
        weather1 = json['weather'][0]['main']
        final = [city, country, temp_kelvin,temp_celsius, weather1]
        return final
    else:
        print("NO Content Found")


def search():
    city = city_text.get()
    weather = getweather(city)
    if weather:
        location_lbl['text'] = "City: "+'{} '.format(weather[0])
        weather_l['text'] = "Conditions: "+weather[4]
        temperature_label['text'] = "Temperature (C): "+str(round(weather[3]))
        
    else:
        messagebox.showerror('Error', "Cannot find {}".format(city))


app = Toplevel()
app.geometry("500x300")


FILENAME = "tree.jpg"
canvas = tk.Canvas(app, width=500, height=400)
canvas.pack()
tk_img = ImageTk.PhotoImage(file = FILENAME)
canvas.create_image(125, 125, image=tk_img)


city_text = StringVar()
city_entry = tk.Entry(app, width=20,textvariable=city_text,highlightthickness=5)
city_entry.config(highlightbackground = "lightblue", highlightcolor= "lightblue")
city_entry.pack()

Search_btn = tk.Button(app, text="Search Weather",width=12, command=search)
Search_btn.pack()

location_lbl = Label(app)
location_lbl.pack()

weather_l = Label(app, text="")
weather_l.pack()

temperature_label = Label(app, text="")
temperature_label.pack()

city_entry_window = canvas.create_window(120, 30, anchor='nw', window=city_entry)
search_button_window = canvas.create_window(260, 30, anchor='nw', window=Search_btn)
location_window = canvas.create_window(240,120,anchor = 'center',window=location_lbl,width=250,height=30)
weather_window = canvas.create_window(240,140,anchor = 'center',window=weather_l,width=250)
temperature_window = canvas.create_window(240,160,anchor = 'center',window=temperature_label,width=250,height=30)

app.mainloop()
```
